```python
xy=d.drop("veil-type",axis=1)# since it has only one value so its absence would not hamper the model
xy
features=list(xy.columns)
```

```python
c=0
for i in range(len(d)):
    if d.loc[i,"stalk-surface-above-ring"] != d.loc[i,"stalk-surface-below-ring"]:
        c=1
        print("Not Same")
        break
if c==0:
    print("Same")
c=0
for i in range(len(d)):
    if d.loc[i,"stalk-color-above-ring"] != d.loc[i,"stalk-color-below-ring"]:
        c=1
        print("Not Same")
        break
if c==0:
    print("Same")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in range(len(features)):
    xy[features[i]]=le.fit_transform(xy[features[i]])
```

```python
xy
```

```python
x=xy.drop("class",axis=1)
```

```python
y=xy.loc[:,["class"]]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclf=DecisionTreeClassifier()
dtclf.fit(x_train,y_train)
dtclf.score(x_test,y_test)
```

```python
import joblib
joblib.dump(dtclf,'dtmodel.pkl')
```
